```python
df = quandl.get("EURONEXT/ABN.4",start_date = '2019-04-01', end_date = '2020-04-01')
```

```python
df.tail()
```

```python
df_daily_return = df.pct_change(periods = 1)
df_3day_return = df.pct_change(periods = 3)
df_7day_return = df.pct_change(periods = 7)
```

```python
fig, axes = plt.subplots(3,1, figsize = (15,10))
sns.set(style="whitegrid")
sns.lineplot(x = df_daily_return.index, y = "Last", data = df_daily_return, alpha = 0.7 ,ax = axes[0],label = '1-day')

sns.lineplot(x = df_3day_return.index, y = "Last", data = df_3day_return, 
             color = "green", alpha = 0.7 ,ax = axes[1],label = '3-day')
sns.lineplot(x = df_7day_return.index, y = "Last", data = df_7day_return,
             color = "red", alpha = 0.7 ,ax = axes[2],label = '7-day')
for i in range(3):
    ax[i].legend()
plt.suptitle("EURONEXT/ABN Return \nFrom 2019-04 to 2020-04",fontsize = 20)
plt.show()
```

```python
df = quandl.get(["EURONEXT/ABN.4","EURONEXT/SANTA.4","EURONEXT/KA.4"],
                start_date = '2016-01-01', end_date = '2017-12-31',collapse = "daily")
```

```python
df.plot(figsize = (15,10))
plt.title("Multiple EURONEXT")
```

```python
df.corr()
```

```python
df = df['EURONEXT/ABN - Last']
```

```python
series_short = df.rolling(window = 5, min_periods = 5).mean()
series_long = df.rolling(window = 30, min_periods = 30).mean()
fig = plt.figure()
df_sma = pd.DataFrame(columns = ['short','long'])
df_sma['short'] = series_short
df_sma['long'] = series_long
df_sma = df_sma.asfreq('D',method = 'ffill')#.dropna()
df_sma.plot(figsize = (15,10))
```

```python
series_short = df.ewm(span = 5).mean()
series_long = df.ewm(span = 30).mean()
fig = plt.figure()
df_sma = pd.DataFrame(columns = ['short','long'])
df_sma['short'] = series_short
df_sma['long'] = series_long
df_sma = df_sma.asfreq('D',method = 'ffill')#.dropna()
df_sma.plot(figsize = (15,10))
```
